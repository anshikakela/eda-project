```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
imdb_data = pd.read_csv('imdb_top_1000.csv')
```

```python
print(imdb_data.info())
```

```python
print(imdb_data.head())
```

```python
print(imdb_data.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(imdb_data['IMDB_Rating'], bins=30, kde=True)
plt.title('Distribution of Ratings in IMDb Top 1000')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='IMDB_Rating', y='Runtime', data=imdb_data)
plt.title('Rating vs Runtime in IMDb Top 1000')
plt.xlabel('Rating')
plt.ylabel('Runtime (minutes)')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Genre', y='IMDB_Rating', data=imdb_data)
plt.title('Rating by Genre in IMDb Top 1000')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```
